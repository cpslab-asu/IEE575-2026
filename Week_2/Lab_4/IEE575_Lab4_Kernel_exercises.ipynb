## Common Setup

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x, x)
```

## 1. Squared Exponential (RBF) Kernel

### Formula

$$k(x, x') = \sigma^2 \exp\left(-\frac{(x - x')^2}{2\ell^2}\right)$$

### Task 1 — Implement the kernel

```python
def se_kernel(X1, X2, lengthscale=1.0, sigma=1.0):

    return sigma**2 * np.exp(-(X1 - X2)**2 / (2 * lengthscale**2))
```

### Task 2 — Plot the kernel matrix

```python
K = se_kernel(X1, X2)

plt.imshow(K, extent=[-5,5,-5,5])
plt.title("1. Squared Exponential (RBF) Kernel")
plt.colorbar()
plt.show()
```

### Task 3 — Explore parameters

### Task 4 — Slice view

```python
x_fixed = 0.0
k_slice = se_kernel(x.reshape(-1, 1), np.array([[x_fixed]])).flatten()

plt.plot(x, k_slice)
plt.title("k(x, x'=0)")
plt.show()
```
